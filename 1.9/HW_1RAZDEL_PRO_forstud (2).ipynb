# Итоговый проект.

Анализ личности клиента — это подробный анализ клиентов компании. Он помогает компании лучше понять своих клиентов и облегчает изменение продукции в соответствии со специфическими потребностями, поведением различных типов клиентов.

Анализ личности клиента помогает компании модифицировать свой продукт в соответствии с целевыми потребителями из различных сегментов потребителей. Например, вместо того, чтобы тратить деньги на маркетинг нового продукта для каждого клиента в базе данных компании, компания может проанализировать, какой сегмент клиентов с наибольшей вероятностью купит продукт, и затем продвигать продукт только для этого сегмента.

3. Тратят ли клиент больше денег, если у него много детей и иждивенцев?


```python
# Создаем новый столбец, обозначающий общее количество детей и иждивенцев
df['Total_Children'] = df['Kidhome'] + df['Teenhome']
df.T
```

 Проверяем, тратят ли клиенты больше денег, если у них больше детей и иждивенцев

```python
for col in ['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']:
    corr, _ = stats.pearsonr(df['Total_Children'], df[col])
    print(f'Корреляция между Total_Children и {col}: {corr}')

mean_expenses_by_children = df.groupby('Total_Children')\
    [['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']].mean()
print(mean_expenses_by_children)
```

На данной диаграмме представлено потребление различных продуктов в зависимости от количества детей. Можно заметить, что с увеличением количества детей, потребление продуктов снижается. 

- Потребление вина (MntWines) достаточно существенно уменьшается при увеличении количества детей.
- Потребление фруктов (MntFruits) также снижается, но не так заметно как потребление вина.
- Потребление мясных продуктов (MntMeatProducts) и рыбы (MntFishProducts) также уменьшается при увеличении количества детей, но не так значительно как потребление других продуктов.
- Потребление сладких продуктов (MntSweetProducts) и золотых продуктов (MntGoldProds) сильно заметно уменьшается с увеличением количества детей.

Эти данные позволяют сделать вывод, что с увеличением количества детей уровень потребления различных категорий продуктов снижается, вероятно из-за финансовой нагрузки, которая возникает при уходе за большим количеством детей.

8. Люди, принявшие предложение о покупке лишь в последней компании тратят больше, чем люди, принявшие сразу предложение первой компании?

Для начала разделим данные на две группы: людей, принявших предложение о покупке только в последней компании (AcceptedCmp3, AcceptedCmp4, AcceptedCmp5, AcceptedCmp1, AcceptedCmp2 все равны 0), и людей, принявших предложение сразу в первой компании (AcceptedCmp1 равен 1).

Далее посмотрим на средние значения потраченных средств на различные виды продуктов для этих двух групп:


```python
df['Total_Purchase'] = df['MntWines'] + df['MntFruits'] + df['MntMeatProducts'] + df['MntFishProducts'] + df['MntSweetProducts'] + df['MntGoldProds']

last_campaign = df[(df['AcceptedCmp3'] == 0) & (df['AcceptedCmp4'] == 0) & (df['AcceptedCmp5'] == 0) & (df['AcceptedCmp1'] == 0) & (df['AcceptedCmp2'] == 0)]
first_campaign = df[(df['AcceptedCmp1'] == 1)]

mean_last_campaign = last_campaign['Total_Purchase'].mean()
mean_first_campaign = first_campaign['Total_Purchase'].mean()

print("Средние значения потраченных средств для групп:")
print("Люди, принявшие предложение только в последней кампании:", mean_last_campaign)
print("Люди, принявшие предложение в первой кампании:", mean_first_campaign)
```

Далее построим корреляционную матрицу для обеих групп:

```python
last_correlation = last_campaign[['Total_Purchase', 'MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']].corr()
first_correlation = first_campaign[['Total_Purchase', 'MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']].corr()

print("Корреляционная матрица для группы, принявшей предложение только в последней кампании:")
print(last_correlation)

print("Корреляционная матрица для группы, принявшей предложение в первой кампании:")
print(first_correlation)
```

Наконец, построим графики потраченных средств для каждой категории продуктов для обеих групп и сделаем выводы:

```python
plt.figure(figsize=(10, 5))
plt.bar(['Wines', 'Fruits', 'MeatProducts', 'FishProducts', 'SweetProducts', 'GoldProds'], last_campaign[['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']].mean(), label='Last Campaign')
plt.bar(['Wines', 'Fruits', 'MeatProducts', 'FishProducts', 'SweetProducts', 'GoldProds'], first_campaign[['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']].mean(), label='First Campaign', alpha=0.5)
plt.legend()
plt.ylabel('Average Amount Spent')
plt.title('Comparison of Average Amount Spent on Different Products')
plt.show()
```

Из графиков и корреляционной матрицы можно сделать вывод, что люди, принявшие предложение только в последней кампании, тратят больше на все виды продуктов по сравнению с людьми, принявшими предложение сразу в первой кампании. Разница в потраченных средствах особенно заметна на категориях вина и мясных продуктов.
