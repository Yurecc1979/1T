```python
# Transform с лямбда-функцией

df.groupby("Category")['Rating'].transform(lambda x: x.mean() - x.std())
```

```python
# Melt позволяет переворачивать таблицу, преобразуя столбцы в строки
pd.melt(df, id_vars=["App"], value_vars=['Reviews'])
```
